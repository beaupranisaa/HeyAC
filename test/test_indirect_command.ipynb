```python
from hey_ac import HeyAC
import yaml
```

```python
h = HeyAC(grammar_path = '../grammars/grammar_indirect_command.txt')
```

```python
texts = [
    'the speed of the fan is too low',
    'fan speed is low',
    'temperature is high',
    'humidity feels too high',
    'The room is very damp',
    'It is burning',
    'it is so cold',
    'the fan is not quite fast',
    'i feel very hot',
    'i am freezing',
    'it is freezing',
    'it is dry in here',
    'the dog feels cold',
    'the ac is very weak',
    'the volume of air is so low',
    'the ac is not running',
    'the flap is not moving',
    'the swing is swinging',
    'The air from the ac is burning',
    'the wind is icy',
    'I am feeling very cold',
    'The AC is running hot',
    "it is not cold enough in the room",
    "i am feeling too hot in here"
]
```

```python
for text in texts:
    print("="*50)
    print(text)
    print()
    print(yaml.dump(h.prune(text)))
```

```python
for text in texts:
    print("="*50)
    print(text)
    print(h.parse(text)[0])
```

```python
parse_tree = h.parse(text)
```
